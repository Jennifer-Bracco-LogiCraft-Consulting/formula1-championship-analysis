# **Analisi delle prestazioni nel Campionato Mondiale di Formula 1 (stagione 2008)**

Questo progetto è stato sviluppato con l’obiettivo di analizzare i risultati delle gare della stagione 2008 di Formula 1 e costruire le classifiche finali dei piloti e dei costruttori a partire dai dati disponibili.

L’elaborazione è stata realizzata utilizzando esclusivamente le basi della programmazione in Python, senza l’uso di librerie esterne, implementando da zero la logica di calcolo dei punteggi e di aggregazione dei risultati.

```python
import csv
from os.path import isfile

# Definizione delle funzioni necessarie

SCORES_LIST = [10, 8, 6, 5, 4, 3, 2, 1]

def position_scores(position):
  """
  Funzione per convertire l'ordine di arrivo in relativo punteggio
  """
  if position in range(1, 9):
    points = SCORES_LIST[position-1]
  else:
    points = 0
  return points

def driver_performance(driver_name, data_rows):
  """
  Funzione per calcolare le performance del pilota indicato dall'utente
  """
  name_found = False

  sum_points = 0
  count_wins = 0
  count_podiums = 0

  for row in data_rows:
    if row["Driver"].upper()==driver_name:
      name_found = True
      position = int(row["Position"])
      points = position_scores(position)
      sum_points+=points

      if position==1:
        count_wins+=1

      if 1<=position<=3:
        count_podiums+=1

  if not name_found:
    raise ValueError("Attenzione! Pilota non trovato.")

  return [sum_points, count_wins, count_podiums]

def aggregate_scores(data_rows, column):
  """
  Funzione per sommare i punteggi e creare i dizionari in base alla colonna del
  dataset indicata come argomento durante la chiamata alla funzione
  (es. Driver o Team).
  """
  dict_of_points = {}

  for row in data_rows:
    key = row[column]
    position = int(row["Position"])
    points = position_scores(position)

    if key not in dict_of_points:
      dict_of_points[key] = 0

    dict_of_points[key]+=points

  return dict_of_points

# per rendere il programma riutilizzabile anche per altri file/anni
while True:
  try:
    file_path = input("Inserire il percorso del file: ")
    assert(isfile(file_path)), f"Attenzione! Il percorso {file_path} non è stato trovato."
    break
  except AssertionError as e:
    print(e)

while True:
  try:
    year = int(input("Inserire l'anno del campionato: "))
    break
  except ValueError:
    print("Attenzione! Inserire l'anno in numeri.")


with open(file_path) as csv_file:
  csv_reader = csv.DictReader(csv_file)
  # variabile creata per poter riutilizzare il dataset nelle varie funzioni senza esaurirsi come csv_reader
  data_rows = list(csv_reader)

  driver_name = input("Inserire il nome del pilota: ").upper()

  try:
    # la funzione di per sé ritorna esattamente quanto richiesto dalle specifiche
    driver_stats = driver_performance(driver_name, data_rows)
    # per una maggiore precisione e leggibilità dell'output ho aggiunto questi print
    total_point, wins, podiums = driver_stats
    print(f"{driver_name} - Totale dei punti: {total_point}; "+
          f"Numero di vittorie: {wins}; Numero di podi: {podiums}")
  except ValueError as e:
    print(e)

  # chiamata alla funzione per il dizionario dei piloti
  drivers_unsorted = aggregate_scores(data_rows, "Driver")
  drivers_standing = sorted(drivers_unsorted.items(), key=lambda x:x[1], reverse=True)
  # per una maggiore precisione e leggibilità dell'output ho aggiunto questo print
  print(f"Per la classifica dei piloti vedere il file: Drivers_Standings_{year}.txt")

  # esportazione classifica piloti in file .txt
  with open(f"Drivers_Standings_{year}.txt", "w") as drivers_standings_file:
    drivers_standings_file.write(f"Drivers Standings {year} Formula 1\n")

    for driver, points in drivers_standing:
      drivers_standings_file.write(f"{driver}: {points}\n")

  # chiamata alla funzione per il dizionario dei team
  teams_unsorted = aggregate_scores(data_rows, "Team")
  teams_standing = sorted(teams_unsorted.items(), key=lambda x:x[1], reverse=True)
  print("Classifica Team:", dict(teams_standing))
```
